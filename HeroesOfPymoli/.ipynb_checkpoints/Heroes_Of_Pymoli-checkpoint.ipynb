```python
# Dependencies and Setup
import pandas as pd

# File to Load
csvfile = "purchase_data.csv"

# Read Purchasing File and store into Pandas data frame
purchase_df = pd.read_csv(csvfile)
```

```python
# Player count- Display the total number of players
total_players = len(purchase_df["SN"].unique())
total_players_df = pd.DataFrame({
    "Total Players":[total_players]
})
total_players_df
```

```python
#Purchasing Analysis (Total)
#Run basic calculations to obtain number of unique items, average price, etc.
#Create a summary data frame to hold the results; Optional: give the displayed data cleaner formatting
unique_items = len(purchase_df["Item ID"].unique())
average_price = round((purchase_df["Price"].mean()), 2)
number_purchases = len(purchase_df["Price"])
total_revenue = purchase_df["Price"].sum()

purchase_analysis_df = pd.DataFrame({
    "Number of Unique Items": [unique_items],
    "Average Price": ["$" + str(average_price)],
    "Number of Purchases": [number_purchases],
    "Total Revenue": ["$" + str(total_revenue)]
})
purchase_analysis_df
```

```python
#Gender Demographics
#Percentage&Count of Male Players, Percentage&Count of Female Players, Percentage&Count of Other/Non-Disclosed
all_genders = []
all_sn = []
unique_genders = []

for gender in purchase_df["Gender"]:
    all_genders.append(gender)

for sn in purchase_df["SN"]:
    all_sn.append(sn)

unique_players = purchase_df["SN"].unique()
unique_players

for i in unique_players:
    unique_player_index = all_sn.index(i)
    unique_genders.append(all_genders[unique_player_index])

gender_count_series = pd.Series(i[0] for i in unique_genders)

male_count = gender_count_series.value_counts()[0]
female_count = gender_count_series.value_counts()[1]
other_count = gender_count_series.value_counts()[2]
male_percent = round(((male_count / total_players) * 100), 2)
female_percent = round(((female_count / total_players) * 100), 2)
other_percent = round(((other_count / total_players) * 100), 2)

gender_count_df = pd.DataFrame({
    "Total Count": [male_count, female_count, other_count],
    "Percentage of Players": [str(male_percent) + "%", str(female_percent) + "%", str(other_percent) + "%"]},
    index = ['Male', 'Female', 'Other / Non-Disclosed']
)
gender_count_df
```

```python
#Age Demographics
#Establish bins for ages, Categorize the existing players using the age bins. Hint: use pd.cut()
#Calculate the numbers and percentages by age group, Create a summary data frame to hold the results
#Optional: round the percentage column to two decimal points

all_ages = []
unique_ages = []

for age in purchase_df["Age"]:
    all_ages.append(age)

for i in unique_players:
    unique_player_index = all_sn.index(i)
    unique_ages.append(int(all_ages[unique_player_index]))

less10 = 0
btw10a14 = 0
btw15a19 = 0
btw20a24 = 0
btw25a29 = 0
btw30a34 = 0
btw35a39 = 0
above40 = 0

for i in range(len(unique_ages)):
    if unique_ages[i] < 10:
        less10 += 1
    elif unique_ages[i] >= 10 and unique_ages[i] <= 14:
        btw10a14 += 1
    elif unique_ages[i] >= 15 and unique_ages[i] <= 19:
        btw15a19 += 1
    elif unique_ages[i] >= 20 and unique_ages[i] <= 24:
        btw20a24 += 1
    elif unique_ages[i] >= 25 and unique_ages[i] <= 29:
        btw25a29 += 1
    elif unique_ages[i] >= 30 and unique_ages[i] <= 34:
        btw30a34 += 1
    elif unique_ages[i] >= 35 and unique_ages[i] <= 39:
        btw35a39 += 1
    else:
        above40 += 1

less10percent = round(((less10 / len(unique_ages)) * 100), 2)
btw10a14percent = round(((btw10a14 / len(unique_ages)) * 100), 2)
btw15a19percent = round(((btw15a19 / len(unique_ages)) * 100), 2)
btw20a24percent = round(((btw20a24 / len(unique_ages)) * 100), 2)
btw25a29percent = round(((btw25a29 / len(unique_ages)) * 100), 2)
btw30a34percent = round(((btw30a34 / len(unique_ages)) * 100), 2)
btw35a39percent = round(((btw35a39 / len(unique_ages)) * 100), 2)
above40percent = round(((above40 / len(unique_ages)) * 100), 2)

age_anlysis_df = pd.DataFrame({
    "Total Count": [less10, btw10a14, btw15a19, btw20a24, btw25a29, btw30a34, btw35a39, above40],
    "Percentage of Players": [str(less10percent) + "%", str(btw10a14percent) + "%", str(btw15a19percent) + "%", str(btw20a24percent) + "%", str(btw25a29percent) + "%", str(btw30a34percent) + "%", str(btw35a39percent) + "%", str(above40percent) + "%"]},
    index = ['<10', '10-14', '15-19', '20-24', '25-29', '30-34', '35-39', '40+']
)

age_anlysis_df
```
